```python
# so, we have to convert the string into list first, then extract the genres as intended
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# we only need top 3 actors and the name of the director
movies['cast'][0]
def convertcast(obj):
    L = []
    obj = ast.literal_eval(obj)
    counter = 0
    for i in obj:
        if counter==3: break
        L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convertcast)
```

```python
movies.head()
```

```python
movies['crew'][0]
def convertCrew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            L.append(i['name'])
            return L
```

```python
movies['crew'] = movies['crew'].apply(convertCrew)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# now, we have to remove the spcae between cast's first and last name
# as if didnt do that, for eg. in the models, two tags will be created for 'Chris Hemsworth'-->'Chris','Hemsworth'
# so now, if any person likes the movies of CH, the movies of 'Chris Evans' will also get recommended, as both have the 'Chris' tag in them
# WE DON'T want this, so we make separate tags as--> 'ChrisHemsworth', 'ChrisEvans' to avoid the mixup

movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

## Now we have to do the VECTORIZATION of tags

so basically, we have to convert text(tags) to vectors
this is called text-vectorization
Every movie will have its own corresponding vector,and we'll recommend the movies which have vectors nearby
the movie user watched

https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english')# stop_words attribute will remove the words of english language(like as, in, the, a etc.)
```

note that, here the max_features=5000 means that we have chosen top 5000 words from the dataset for the vectorization of movies(excluding english words)

```python
vectors = cv.fit_transform(new_df['tags']).toarray()## by default, it is a sparse matx., we explictly convert it to numpy array
```

```python
print(vectors)
print(vectors.shape)
```

```python
pd.DataFrame(cv.get_feature_names_out().tolist()).head(50)
```

we can notice a problem here, both words ability and abilities, actor and actors,etc. as both means the same but are listed differently..

```python
# ['loving','love','loved'] --> ['love','love','love']
import nltk # natural processing lib.
# basically it will convert the word into its ROOT-word
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

in higher dimensions , euclidean distance fails
so instead, we use cosine distance b/w movie vectors as a form of comparision

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    # note that by sorting distances, we will lose the index position of the movies in the similarity matx,
    # thats why we will use enumerate function to store the index values
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```

```python
recommend('The Avengers')
```

```python
recommend('Just Married')
```

```python
recommend('Batman Begins')
```
